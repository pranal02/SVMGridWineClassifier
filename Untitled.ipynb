```python
from sklearn.datasets import load_wine
```

```python
wine=load_wine()
```

```python
print(wine)
```

```python
print(wine.DESCR)
```

```python
x = wine.data
y=wine.target
features=wine.feature_names
```

```python
import seaborn as sns
import matplotlib.pyplot as plt
for i in range(13):
    sns.boxplot(x=y,y=x[:,i])
    plt.ylabel(features[i])
    plt.show()
```

```python
for i in range(13):
    plt.hist(x[:,i],edgecolor='black')
    plt.title(features[i])
    plt.show()
```

```python
target_name =wine.target_names
print(target_name)
```

```python
from sklearn.svm import SVC
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.3,random_state=32)
```

```python
kernel=['linear','rbf','poly','sigmoid']
for i in kernel:
    model = SVC(kernel=i,C=1.0)
    model.fit(x_train,y_train)
    print("For Kernel",i)
    print("Accuracy is ",model.score(x_test,y_test))
```

```python

for i in range(1,10):
    model = SVC(kernel='poly',degree=i,C=1.0)
    model.fit(x_train,y_train)
    print("Accuracy on test data: \t",model.score(x_test,y_test))
    print("Accuracy on training data : \t",model.score(x_test,y_test))
```

```python
# why to for loop every time there is method called grid search 
from sklearn.model_selection import GridSearchCV
param_grid={'C':[0.1,1,100,1000] , 'kernel':['linear','rbf','poly','sigmoid'],'degree':[1,2,3,4,5,6]}
grid=GridSearchCV(SVC(),param_grid)
grid.fit(x_train,y_train)
```

```python
print(grid.best_params_)
print(grid.score(x_test,y_test))
```

```python
# we conclude that SVM with kernel liner with c = 1 and degree =1 is the best model for prediction
```
